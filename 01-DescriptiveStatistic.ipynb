### 1.1 數據集資料轉換至Pandas DataFrame (Convert a Scikit-learn dataset to a Pandas dataset)

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston

bostaon= load_boston()
data = pd.DataFrame(bostaon.data, columns=[bostaon.feature_names])
data['Target'] = pd.Series(data=bostaon.target, index=data.index)
```

### 1.2 觀察資料集中心位置 -- （均值、中位數、衆數）

```python
#利用 Pandas's 現有function 針對全體資料進行描述性統計分析
print("----mean() function----")
print(data.mean()) #取平均值
print("----median() function----")
print(data.median()) #取中位數值
print("----mode() function----")
print(data.mode()) #取眾數
```

### 1.3 觀察資料集離散程度--（極差、方差、標準差、變異係數）

指標|英文|說明 | 公式
---------|---------|---------|---------
極差 (全距)|Range|最大值與最小值之間的差額 | ![Alt text](https://wikimedia.org/api/rest_v1/media/math/render/svg/6ca934851e73180471a048503c22b497696e7222)
變異數 (方差)|Variance|各個數據與其算術平均數的離差平方和的平均數|![Alt text](https://wiki.mbalib.com/w/images/math/1/a/2/1a28eba1b3ca17b3cc16055d9ef32662.png)
標準差 (標準偏差、均方差)| Standard Deviation |為變異數開平方根，反映組內個體間的離散程度 | ![Alt text](https://wikimedia.org/api/rest_v1/media/math/render/svg/7b165a48481efe07c6c6430c8d2e86f8e723da9f)
變異係數|Coefficient of Variation|資料的標準差除以平均數所得的商化為百分比所得之值|![Alt text](http://estat.ncku.edu.tw/topic/desc_stat/base/cv.htm4.gif)


```python
print("----Range : max()-min()--------")
print(data.max()-data.min()) # 極差 (全距)
print("----var() function----")
print(data.var()) #取方差
print("----std() function----")
print(data.std()) #取標準差

print("----coefficient of variation----")
print(data.std()/data.mean()) #取變異係數
```

### 1.4 觀察資料集偏差程度--（z-分數）

Z值代表原始分數和母體平均值之間距離，以標準差為單位計算。
* 在原始分數低於平均值時Z則為負數，反之則為正數。=> 點到均值之間有多少個標準差
* 公式 : (原始分數-母體平均值) / 母體標準差

![Alt text](https://wikimedia.org/api/rest_v1/media/math/render/svg/5ceed701c4042bb34618535c9a902ca1a937a351)


```python
df_zscore = (data - data.mean())/data.std()
df_zscore
```

### 1.5 觀察資料集相關程度--（協方差、相關係數）

* 協方差 : 用於衡量兩個變數的總體誤差<br>
兩個變數的變化趨勢相反，即其中一個大於自身的期望值，另外一個卻小於自身的期望值，那麼兩個變數之間的協方差就是負值

* 相關係數是研究變數之間線性相關程度的量 (相關係數0.3以下為低相關，0.3~0.7為中等相關，0.7以上為高度相關)
  <br> r = 1，完全相關 
　<br> 0 < r < 1，正相關 
　<br> r = 0，無關 
　<br> -1 < r < 0，負相關 
　<br> r = -1，完全負相關 

```python
# 計算協方差
data.cov()
```

```python
# 計算相關係數
data.corr()
```

```python
#計算變數之間的線性相關係數
data[['CRIM','ZN','INDUS']].corr()
```
